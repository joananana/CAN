```python
import numpy as np
import matplotlib.pyplot as plt

def corrente(_, i):
  V = 100
  R = 10
  L = 1
  return (1/L) * (V - R * i)

def range_kutta4(function,    # função com argumentos f(x, y)
                 x_inicial,   # valor incial de x
                 x_final,     # valor final de x
                 y_inicial,   # valor incial de y
                 n_intervalos):   # quantidade de intervalos/ resolução
  x = [None for i in range (n_intervalos+1)]
  y = [None for i in range (n_intervalos+1)]
  x[0] = x_inicial
  y[0] = y_inicial
  h = (x_final-x_inicial) / n_intervalos # distância entre os passos

  for p in range(n_intervalos):
    k1 = function(x[p],     y[p])
    k2 = function(x[p]+h/2, y[p]+h/2*k1)
    k3 = function(x[p]+h/2, y[p]+h/2*k2)
    k4 = function(x[p]+h,   y[p]+h  *k3)

    x[p + 1] = x_inicial + (p+1)*h
    y[p + 1] = y[p] + (1/6) * (k1 + 2*k2 + 2*k3 +k4)*h
  return x, y


t_inicial = 0
t_final = 1
i_inicial = 0
n_intervalos = 1000

t, i = range_kutta4(corrente, t_inicial, t_final, i_inicial, n_intervalos)
print(f'Corrente no tempo {t[-1]}s é de {i[-1]}A')

# Visualização
plt.plot (t, i, linewidth =2)
plt.grid(True)
plt.xlabel('Tempos (s)')
plt.ylabel('Corrente (A) ')
plt.show()
```
