```python
## Joana Wasserberg
## Cálculo Numérico
## Encontrar a raiz da função de um retificador de meia onda com carga RL
## 04/09/2023

import matplotlib.pyplot as plt
import numpy as np

############### Método Bissecção ###############
def bisection(  a,                  # inicial
                b,                  # final
                iterations=100,     # número máximo de iterações
                tolerance=0.0001    # erro tolerado
                ):


    ##### Plot da função ###############
    steps = 100
    plt.figure(dpi=150)
    plt.grid()
    x = np.linspace(a, b, steps)    # A to B in N steps
    plt.plot(x,f(x),'b-')
    ##########################

    counter = 0
    while ((b-a) > tolerance) & (counter < iterations):
    #a nunca será maior do que b

        counter += 1
        x0 = (b+a)/2

        if f(x0) * f(b) > 0:    # New interval: [a, x0]
            b = x0
        else:                   # New interval: [x0, b]
            a = x0

        print('[',a,',',b,']', 'Interactions:',counter)

    plt.plot(x0, f(x0), 'ro')
    plt.text( x0 , f(x0) , '$x'+ str(counter) +'$', ha='left', va='top' )
    return x0


# Função i(b) #
def f(  B,                  # Ângulo de extinção
        Im = 6.78,          # Corrente Média
        Tet = 0.646,        # Ângulo de disparo
        sinAlfTet = 2.67,   # Seno de alfa - teta
        wT = 0.754          # Constante de tempo
      ):

    return Im*(np.sin(B-Tet)-sinAlfTet*np.exp((-B)/wT))

print('Result:', bisection(0.785, 6.28, 20, 0.0001 ))
```
