```python
## Joana Wasserberg
## Cálculo Numérico
## Encontrar a curva de resistência de um fio
## 19/11/2023

import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import splev, splrep, CubicSpline

x = np.array([1.5,2.,2.2,3.,3.8,4.])
y = np.array([4.9,3.3,3.,2.,1.75,1.5])
x_graph = np.linspace(min(x), max(x), 1000)

yy= 1.75
n = 4
P = np.polyfit(x, y, n)
s = np.polyval(P, yy)
print(f'Polinômio grau {n} \nValor de f(x) em x = 1.75: {s:.4f}\n')

n = 3
P = np.polyfit(x, y, n)
s = np.polyval(P, yy)
print(f'Polinômio grau {n} \nValor de f(x) em x = 1.75: {s:.4f}\n')

f = CubicSpline(x,y, bc_type = 'natural')
y_n = f(x_graph)
print(f'Spline_cubic(1.75) = {f(1.75):.4f}')
plt.plot(x_graph, y_n, '-b', label = "CubicSpline")

plt.grid()
plt.plot(x, y, "om", label = "f(x)")
plt.title('Cubic Spline Interpolation')
plt.legend()
plt.show()
```
